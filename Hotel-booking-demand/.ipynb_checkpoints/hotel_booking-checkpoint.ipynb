```python
#处理缺失值
df.isnull().sum()[df.isnull().sum()>0]
```

```python
# children字段中是数值型变量，且偏态分布，即用中位数替代
df["children"].fillna(df["children"].median(), inplace=True)
#country字段是类别变量，用众数填充
df['country'].fillna(df['country'].mode()[0],inplace=True)
#agent空值用0填充，表示没有代理
df['agent'].fillna(0,inplace=True)
#company缺失值太多，删除变量
del df['company']
```

```python
#查看非数值变量,是否有格式错乱或者需要数据处理的
for c in df.columns:
    if c == 'reservation_status_date':
        continue
    if df[c].dtype == object:
        print(c,df[c].unique().tolist())
```

```python
#数据处理
# 初始条件，餐饮字段中的Undefined / SC –无餐套餐为一类
df["meal"].replace("Undefined", "SC", inplace=True)

#arrival_date_month 英文字符转为数字格式
df['arrival_date_month'] = df['arrival_date_month'].map((list(calendar.month_name)).index)
```

```python
#删除异常值
df.drop(df[df['adults']+df['children']==0].index,inplace=True)
df.drop(df[df['adr']<0].index,inplace=True)
```

# 酒店运营情况分析
目标：对城市酒店和度假酒店对比分析其运营情况
1. 酒店预定量、入住量、取消率
2. 酒店人均价格

```python
df_ = df.groupby(['hotel','is_canceled'])['lead_time'].count()

def func(pct,data):
    b = int(pct*data/100.)
    return "{:.2f}%\n{:d}".format(pct,b)

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["font.serif"] = ["SimHei"]
fig = plt.figure(figsize=(6,5))

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(2)*4)
inner_colors = cmap([1, 2, 5, 6])

ax1 = fig.add_subplot(1,1,1)
w1, t1, at1 = ax1.pie([df_.loc['City Hotel'].sum(),df_.loc['Resort Hotel'].sum()],radius=1,wedgeprops=dict(width=0.4, edgecolor='w'),
        pctdistance=0.75,colors=outer_colors,autopct=lambda pct:func(pct,df_.sum()))
w2, t2, at2 = ax1.pie(df_,radius=0.6,wedgeprops=dict(width=0.4, edgecolor='w'),
        colors=inner_colors,autopct=lambda pct:func(pct,df_.sum()))
ax1.set_title('2015-2017年度不同酒店预定情况', fontsize=16)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.42)
kw = dict(arrowprops=dict(arrowstyle="-", color="k"), bbox=bbox_props, zorder=0.1, va="center")
for i,p in enumerate(w2):
    text = ["城市酒店-入住", "城市酒店-取消", "度假酒店-入住", "度假酒店-取消"]
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle, angleA=0, angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    plt.annotate(text[i], xy=(x, y), xytext=(1.15*np.sign(x), 1.2*y),
                horizontalalignment=horizontalalignment, **kw, fontsize=12)
    
    plt.legend(w1, ["城市酒店", "度假酒店"], loc="upper right", bbox_to_anchor=(0, 0, 0.2, 1), fontsize=11)

plt.show()
```

1. 城市酒店单量比度假酒店多
2. 城市酒店的取消率高于度假酒店

```python
#按照月份看每个酒店的预定量以及取消率
df_ = df.groupby(['arrival_date_year','arrival_date_month','hotel'])['is_canceled'].agg({'count':'count','canceled':sum}).reset_index()
df_ = df_.groupby(['arrival_date_month','hotel'])['count','canceled'].mean().reset_index()
df_['canceled_rate'] = df_['canceled']/df_['count']
df_['check_in'] = df_['count'] - df_['canceled']
fig = plt.figure(figsize=(10,15))

ax1 = fig.add_subplot(3,1,1)
sns.barplot(data=df_,x='arrival_date_month',y='count',hue='hotel',ax=ax1)
ax1.set_title('每月酒店的预定量')

ax2 = fig.add_subplot(3,1,2)
sns.barplot(data=df_,x='arrival_date_month',y='check_in',hue='hotel',ax=ax2)
ax2.set_title('每月酒店的入住量')

ax3 = fig.add_subplot(3,1,3)
sns.lineplot(data=df_,x='arrival_date_month',y='canceled_rate',hue='hotel',ax=ax3)
ax3.set_title('每月酒店的取消率')
```

1. 度假酒店订单量全年比较稳定，城市酒店预定高峰期在每年的4、5、6、9、10月份
2. 城市酒店全年最忙的月份是3、4、5、6、9、10月份，度假酒店的入住量也会增多--（每年的旅游季）
3. 全年城市酒店的取消率高于度假酒店，4-10月份订单取消率偏高，并且度假酒店取消率变化幅度比城市酒店大（估计是12，1，2月份订单偏少，且为寒假期间，好不容易订个酒店跨年旅游，大多早就计划好，不容易取消）

### 计算酒店入住的人均价格
$perp = \frac{adr}{adults+children}$

```python
df['adr_pp'] = df['adr']/(df['adults']+df['children'])
df_ = df.groupby(['arrival_date_month','hotel'])['adr_pp'].mean().reset_index()

fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(1,1,1)
sns.lineplot(data=df_,x='arrival_date_month',y='adr_pp',hue='hotel')
ax1.set_title('每个酒店的人均价格变化情况')

plt.show()
```

1. 春秋季节虽然人均价格增长，入住量没有减少，确实旅游季的时候酒店都会贵点
2. 度假酒店6-9月份人均价格上升，入住量减少，取消率增加，可能这之间存在因果关系

```python
df_ = df.groupby(['adr_pp','hotel'])['is_canceled'].agg({'count':'count','canceled':sum}).reset_index()
df_['canceled_rate'] = df_['canceled']/df_['count']
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(2,1,1)
sns.barplot(data=df_,x='adr_pp',y='count',hue='hotel',ax=ax1)
ax1.set_title('不同价格酒店的预定量')

ax2 = fig.add_subplot(2,1,2)
sns.barplot(data=df_,x='adr_pp',y='canceled_rate',hue='hotel',ax=ax2)
ax2.set_title('不同价格酒店的取消率')

plt.show()
```
